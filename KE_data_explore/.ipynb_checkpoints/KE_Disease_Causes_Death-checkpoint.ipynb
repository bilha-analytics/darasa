## Adding Other Datasets

```python
db_top_death_causes = pd.read_csv( "%s/KNBS_et_al/national-registered-deaths-by-major-causes-2010-2013.csv" % DATA_DIR)
db_medics = pd.read_csv( "%s/KNBS_et_al/national-registered-medical-personnel-2000-to-2013.csv" % DATA_DIR)

db_medics.drop(10, inplace=True)
db_medics.rename(index=str, columns={"2013*":"2013"}, inplace=True)

db_top_death_causes.tail()
db_medics.tail()
```

```python
dataset["Function"] = "Diseases"
db_medics["Function"]= "Staffing"
db_top_death_causes["Function"] = "Deathers"
```

```python
db_top_death_causes = pd.DataFrame( pd.melt( db_top_death_causes, id_vars=["Function", "Cause"], var_name="Year", 
                                            value_name="Nmb.Deaths", value_vars=["2010", "2011","2012", "2013"]))
db_top_death_causes.sample(3)


db_medics = pd.DataFrame( pd.melt( db_medics, id_vars=["Function", "Type of Personnel Over the Years"], var_name="Year", 
                                            value_name="Nmb.Staff", 
                                            value_vars=["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011","2012", "2013"]))
db_medics.sample(3)

dataset.sample(3)
```

```python
db = pd.merge( pd.merge( db_medics, db_top_death_causes,on="Year", how="outer"), dataset, on=["Year",], how="outer")
db["id"] = db.index

db.head()
```

```python
#db.set_index("Year")
#db.head()
```

```python
def rev(x):
    return np.log2( np.mean(x))

pv = pd.pivot_table(db, index=["Year"], values=["Nmb.Staff", "Nmb.Deaths", "Nmb.Cases"], aggfunc=rev)
pv
pv.plot(kind="line", table=True, figsize=(13,7))
```

```python
print("\n\n 1. Staffing over the years")
pv = pd.pivot_table(db, index=["Type of Personnel Over the Years"], columns=["Year"], values=["Nmb.Staff", ], aggfunc=np.mean)
pv
pv.plot(kind="bar", figsize=(13,7))
```

```python
print("\n\n2. Diseases over the years")
pv = pd.pivot_table(db, index=["Disease Name"], columns=["Year"], values=["Nmb.Cases", ], aggfunc=np.mean)
pv
pv.plot(kind="bar",  figsize=(13,7))
```

```python
print("\n\n3. Death Causes over the years")
pv = pd.pivot_table(db, index=["Cause"], columns=["Year"], values=["Nmb.Deaths", ], aggfunc=np.mean)
pv
pv.plot(kind="bar",  figsize=(13,7)) #.sort_values(["mean","Nmb.Deaths"]) 
```

```python
t,g = plotCorrz( corrz(db, ["Nmb.Staff", "Nmb.Deaths", "Nmb.Cases"]) )
t
```

```python
t,g = plotCorrz( corrz(db[ db["Year"].isin( ["2010", "2011", "2012", "2013"])], ["Nmb.Staff", "Nmb.Deaths", "Nmb.Cases"]) )
t
```

### Another - Append instead of merge

```python
dataset.head(3)
db_medics.head(3)
db_top_death_causes.head(3)
```

```python
db2 = dataset.rename( index=str, columns={"Disease Name":"Description", } ).append(
    db_medics.rename(index=str, columns={"Type of Personnel Over the Years":"Description", "Nmb.Staff":"Nmb.Cases"})
).append( db_top_death_causes.rename(index=str, columns={"Cause":"Description", "Nmb.Deaths":"Nmb.Cases"}))
#help(db.rename)
db2.sample(7)
```

```python
print("\n\n1. Function Numbers over the years")
pd.pivot_table(db2, index=["Year"], columns=["Function"], values=["Nmb.Cases", ], aggfunc=np.sum).round(0)
    
pv = pd.pivot_table(db2, index=["Year"], columns=["Function"], values=["Nmb.Cases", ], aggfunc=rev) 
pv.plot(kind="line",  figsize=(13,7)) #.sort_values(["mean","Nmb.Deaths"]) 

t,g = plotCorrz( corrz(pv, ["Nmb.Cases"]))
t
```

```python
print("\n\n1. Staffing Numbers over the years")
pv = pd.pivot_table(db2[ db2["Function"]=="Staffing"], index=["Description"], columns=["Year"], values=["Nmb.Cases", ], aggfunc=np.mean)
pv
pv.plot(kind="bar",  figsize=(13,7)) 
```

```python
print("\n\n1. Death Numbers over the years")
pv = pd.pivot_table(db2[ db2["Function"]=="Deathers"], index=["Year"], columns=["Description"], values=["Nmb.Cases", ], aggfunc=np.mean)
#pv.T
pv.T.plot(kind="bar",  figsize=(13,7))

len(db2[ db2["Function"]=="Deathers"])

# correlations 
t,g = plotCorrz( corrz(pv, ["Nmb.Cases"]))
t.round(2)
```
